# Minimal Reports

```python
### Udacity_AZDIAS_052018.csv: Demographics data for the general population of Germany; 891 211 persons (rows) x 366 features (columns).
data_dir = '/content/drive/MyDrive/data/2021-05-15_udacity-mle-sagemaker/Bertelsmann-Arvato/'
def generate_report_min(csv_name, report_name):
    path = f"{data_dir}{csv_name}"
    df = pd.read_csv(path, sep=';')
    profile = ProfileReport(df, 
                            title=f"Pandas Profiling Report - {csv_name}", 
                            explorative=True,
                            minimal=True)
    profile.to_file(f"{data_dir}/EDA/{report_name}_min.html")
```

# Full Reports of Columns

```python
file_name = 'Udacity_AZDIAS_052018.csv'
cols_azdias = pd.read_csv(f'{data_dir}{file_name}', sep=';', nrows=1).columns
print(cols_azdias.to_list())
```

https://github.com/pandas-profiling/pandas-profiling/blob/master/src/pandas_profiling/config_default.yaml  
https://stackoverflow.com/questions/56035523/how-to-profile-large-datasets-with-pandas-profiling  

```python
def generate_report_columns(csv_name, report_name, usecols):
    path = f'{data_dir}{csv_name}'
    df = pd.read_csv(path, sep=';', usecols=usecols)
    profile = ProfileReport(df, 
                            title=f'Pandas Profiling Report - {csv_name}', 
                            explorative=True,
                            # correlations={'pearson': {'calculate': False},
                            #               'spearman': {'calculate': False},
                            #               'kendall': {'calculate': False},
                            #               'phi_k': {'calculate': False},
                            #               'cramers': {'calculate': False}}
                            )
    profile.to_file(f"{data_dir}/EDA/{report_name}.html")
```

```python
%%time
usecols = [col  for col in cols_azdias if col[:4]=='D19_']
print(len(usecols)) # 68
generate_report_columns('Udacity_AZDIAS_052018.csv', 'profile_azdias_d19', usecols)
```
